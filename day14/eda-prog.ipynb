```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from scipy.stats import gaussian_kde #libraray for density plot
```

```python
tips = sns.load_dataset("tips")
#loading the dataset
```

```python
tips.describe()
#to understand the data 
```

```python
sns.displot(tips["total_bill"], kde=True)
#to plot the distribution of total bill with a kernel density estimate
```

```python
sns.jointplot(x=tips.tip, y=tips.total_bill)
#to plot the relationship between tip and total bill
```

```python
sns.jointplot(x= tips.tip, y= tips.total_bill, kind='reg')
#to plot the relationship between tip and total bill with a regression line
```

```python
sns.jointplot(x= tips.tip, y= tips.total_bill, kind='hex')
#to plot the relationship between tip and total bill with a hexagonal binning plot
```

```python
sns.pairplot(tips, kind= 'reg')
#to plot pairwise relationships in the dataset with regression lines
```

```python
sns.pairplot(tips, hue= 'time')
#to plot pairwise relationships in the dataset with different colors for different times of day
```

```python
sns.pairplot(tips, hue= 'day')
#to plot pairwise relationships in the dataset with different colors for different days of the week
```

```python
sns.heatmap(tips.corr(numeric_only=True), annot=True)
#to plot a heatmap of the correlation matrix for the numeric columns in the dataset
```

```python
sns.boxplot(x=tips.total_bill)
sns.boxplot(tips.tip)
#to plot boxplots for total bill and tip to identify outliers and understand the distribution of these variables
```

```python
sns.countplot(x=tips.day)
sns.countplot(x=tips.sex)
#to plot countplots for the day of the week and the sex of the customers to understand the distribution of these categorical variables
```

```python
tips.sex.value_counts().plot(kind='pie', autopct='%1.1f%%')
tips.sex.value_counts().plot(kind='bar')
#to plot a pie chart and a bar chart for the distribution of the sex of the customers
```

```python
sns.countplot(data=tips[tips.time == 'Dinner'], x='day')
sns.countplot(data=tips[tips.time == 'Lunch'], x='day')
#to plot countplots for the day of the week for dinner and lunch times to understand the distribution of customers across different days for different times of day
```

```python
fg = sns.FacetGrid(tips, row='smoker', col='time')
fg.map(sns.histplot, 'total_bill')
#to plot histograms of total bill for smokers and non-smokers across different times of day using FacetGrid
```

```python
plt.hist(tips['total_bill'], bins=20)
#to plot a histogram of total bill with 20 bins to understand the distribution of total bill amounts
```

```python
plt.scatter(tips['tip'], tips['total_bill'])
#to plot a scatter plot of tip vs total bill to visualize their relationship
```
